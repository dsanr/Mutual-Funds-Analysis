```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import datetime as dt
```

```python
nifty_500 = pd.read_excel('nifty 500.xlsx', index_col='Date')
nifty_500.describe()
```

```python
nifty_500['1 year returns'] = nifty_500['Total Returns Index'].pct_change(periods=-250)*-1
#rolling(window=3).apply(lambda x : x[1] - x[0])
nifty_500['1 year returns'].head()
```

```python
#nifty_500.style.format({'1 year returns' : '{:.2%}'}).head()
```

```python
nifty_500.describe()
```

```python
plt.figure(figsize=(15,7))
plt.plot(nifty_500.index, nifty_500['P/E'])
```

```python
nifty_500_d = nifty_500.dropna()
```

```python
nifty_500_d.describe()
```

```python
plt.style.use('seaborn-dark')
fig, ax1 = plt.subplots(1,1,figsize=(15,7), dpi = 200)
ax1.plot(nifty_500_d.index, nifty_500_d['1 year returns'], '-g', label='return')
ax1.legend(loc='upper center', fontsize=15)
ax1.set_xlabel('year')
ax2 = ax1.twinx()
ax2.plot(nifty_500.index, nifty_500['P/E'], '-r', label='P/E')
ax2.legend(loc='best', fontsize=15)
```
