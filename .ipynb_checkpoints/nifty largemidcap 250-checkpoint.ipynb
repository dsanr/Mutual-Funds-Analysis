```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import datetime as dt
```

```python
large_mid_250 = pd.read_excel('nifty largemidcap 250.xlsx', index_col='Date')
large_mid_250.describe()
```

```python
large_mid_250['1 year returns'] = large_mid_250['Total Returns Index'].pct_change(periods=-250)*-1
#rolling(window=3).apply(lambda x : x[1] - x[0])
large_mid_250['1 year returns'].head()
```

```python
large_mid_250.describe()
```

```python
plt.figure(figsize=(15,7))
plt.plot(large_mid_250.index, large_mid_250['P/E'], 'r')
plt.twinx()
plt.plot(large_mid_250.index, large_mid_250['Total Returns Index'])
```

```python
large_mid_250_d = large_mid_250.dropna()
```

```python
large_mid_250_d.describe()
```

```python
plt.style.use('seaborn-dark')
fig, ax1 = plt.subplots(1,1,figsize=(15,7), dpi = 200)
ax1.plot(large_mid_250_d.index, large_mid_250_d['1 year returns'], '-g', label='return')
ax1.legend(loc='upper center', fontsize=15)
ax1.set_xlabel('year')
ax2 = ax1.twinx()
ax2.plot(large_mid_250.index, large_mid_250['P/E'], '-r', label='P/E')
ax2.legend(loc='best', fontsize=15)
```
