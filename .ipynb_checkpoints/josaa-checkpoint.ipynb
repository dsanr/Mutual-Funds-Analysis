```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('josaa seats.xlsx', index_col=0)
```

```python
df
```

```python
df.info()
```

```python
df.drop(1972, inplace=True)
```

```python
df['Program-Total  Seat Capacity  Female Supernumerary'].str.split()
```

```python
total = df['Program-Total  Seat Capacity  Female Supernumerary'].str.split(expand=True)
```

```python
df['Program-Total  Seat Capacity'] = total[0]
```

```python
df['Program-Total  Female Supernumerary'] = total[1]
```

```python
df.drop('Program-Total  Seat Capacity  Female Supernumerary', axis=1, inplace=True)
```

```python
df['Total(includes Female Supernumerary)'] = df['Total(includes Female Supernumerary)'].str.split(n=1, expand=True)[0]
```

```python
df.head()
```

```python
cols = df.columns[-3:]
```

```python
df[cols] = df[cols].astype(float)
```

```python
df.info()
```

```python
df.iloc[:, 5:-2].sum()
```

```python
df.iloc[:, 5:-3].sum()/(df['Program-Total  Seat Capacity'].sum()/2)
```

```python
((df['Program-Total  Female Supernumerary'].sum()/2) / df['Total(includes Female Supernumerary)'].sum())
```

```python
(df.iloc[:, 5:-3].sum()/(df['Program-Total  Seat Capacity'].sum()/2))[1::2].sum()
```

```python
df.groupby('Program Name').sum().sort_values('OPEN', ascending=False)
```

```python
(df.groupby('Program Name').sum().sort_values('OPEN', ascending=False).iloc[:,-1] / df.groupby('Program Name').sum().sort_values('OPEN', ascending=False).iloc[:,-2]).sort_values(ascending=False)[:15]
```

```python
df.groupby('Institute Name').sum().sort_values('Total(includes Female Supernumerary)', ascending=False)
```
