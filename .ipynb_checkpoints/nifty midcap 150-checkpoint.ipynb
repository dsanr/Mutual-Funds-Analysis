```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import datetime as dt
```

```python
midcap_150 = pd.read_excel('nifty midcap 150.xlsx', index_col='Date')
midcap_150.describe()
```

```python
midcap_150['1 year returns'] = midcap_150['Total Returns Index'].pct_change(periods=-250)*-1
#rolling(window=3).apply(lambda x : x[1] - x[0])
midcap_150['1 year returns'].head()
```

```python
midcap_150.describe()
```

```python
plt.figure(figsize=(15,7))
plt.plot(midcap_150.index, midcap_150['P/B'], 'r')
plt.twinx()
plt.plot(midcap_150.index, midcap_150['Total Returns Index'])
```

```python
midcap_150_d = midcap_150.dropna()
```

```python
midcap_150_d.describe()
```

```python
plt.style.use('seaborn-dark')
fig, ax1 = plt.subplots(1,1,figsize=(15,7), dpi = 200)
ax1.plot(midcap_150_d.index, midcap_150_d['1 year returns'], '-g', label='return')
ax1.legend(loc='upper center', fontsize=15)
ax1.set_xlabel('year')
ax2 = ax1.twinx()
ax2.plot(midcap_150.index, midcap_150['P/B'], '-r', label='P/E')
ax2.legend(loc='best', fontsize=15)
```
